# **2. Read Dataset**

 Memuat dan mempersiapkan dataset Breast Cancer dari scikit-learn dalam bentuk DataFrame menggunakan pandas.

```python
import pandas as pd
from sklearn import datasets

# Memuat dataset Breast Cancer dari scikit-learn
bc = datasets.load_breast_cancer()

# Mendefinisikan variabel dependen dan variabel target
X = bc.data
y = bc.target

# Mengonversi data fitur dan target menjadi DataFrame
df_X = pd.DataFrame(X, columns=bc.feature_names)
df_y = pd.Series(y, name='Diagnosis')
```

```python
# Menggabungkan fitur dan target dalam satu DataFrame
df = pd.concat([df_X, df_y], axis=1)

df.head(10)
```

## Distribusi Data

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mengambil hanya kolom numerik dari DataFrame
columns_num_tr = df.select_dtypes(include=['float']).columns

# Menyiapkan grid subplot secara otomatis berdasarkan jumlah kolom numerik
n_cols = 3  # Jumlah kolom subplot per baris
n_rows = (len(columns_num_tr) + 1) // n_cols  # Menentukan jumlah baris

fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(8, n_rows * 3))
axes = axes.flatten()

# Membuat boxplot untuk setiap kolom numerik
for i, col in enumerate(columns_num_tr):
    sns.boxplot(y=df[col], ax=axes[i], color='skyblue')
    axes[i].set_title(f'Boxplot of {col}', fontsize=10)
    axes[i].set_ylabel(col, fontsize=8)
    axes[i].grid(False)

# Menghapus subplot kosong jika ada
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
